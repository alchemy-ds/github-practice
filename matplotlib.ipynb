```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,6,8]
fig,ax=plt.subplots()
ax.set_title('My first plot')
ax.set_xlabel('counts')
ax.set_ylabel('frequency')
ax.plot(x,y)
ax.legend("counts vs frequency")
ax.grid(visible=True)
plt.show()     # usually not needed in the jupyter environment
# Exploring the anatomy of the plot
```

```python
np.random.seed(19680801)  # seed the random number generator.
data =pd.DataFrame( {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)})
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
print(data)
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.scatter('a','b',c='c',s='d',data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

```python
np.random.seed(1483)
data1,data2,data3,data4=np.random.randn(4,100)
fig, ax = plt.subplots(figsize=(5, 2.7))
x = np.arange(len(data1))
ax.plot(x, np.cumsum(data1), color='blue', linewidth=3, linestyle='--')
ax.plot(x, np.cumsum(data2), color='orange', linewidth=2,marker='o')
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.scatter(data1, data2, s=50, facecolor='C0', edgecolor='k',hatch='+')
```

```python
fig,ax=plt.subplots()
ax.boxplot(data1,notch=True,whis=True,meanline=True,showmeans=True,showfliers=True,showbox=True)
ax.annotate(f'Mean={np.mean(data1):.2f}',xy=(1,np.mean(data1)))


plt.show()
```

```python
fig, axs = plt.subplots(3, 3, figsize=(10, 5.7), layout='constrained')
xdata = np.arange(len(data1))  # make an ordinal for this
data = 10**data1
axs[0,0].scatter(xdata, data)
fig[0,1].subplots(2,2)
```
